```python
'''
A vantagem de utilizar a Scikit-Learn é que não precisamos de fato implementar o algoritmo, pois já está implementado e ainda de maneira bem performática.
O que faremos é a execução dos seus métodos, ajuste dos parâmetros e principalmente, entender os resultados.
Usaremos como dados de teste o dataset iris, de qual é bem conhecido no mundo Data Science.
Como nosso objetivo é estudar o funcionamento do algoritmo K-Means, este dataset atende bem o propósito.
'''

import pandas as pd
iris = pd.read_csv("https://gist.github.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv")
iris.head()
```

```python
'''
Iremos utilizar apenas os valores dos atributos SepalLength, SepalWidth, PetalLength e PetalWidth.
Para isso vou armazenar os valores na variável X, veja com o comando abaixo.
'''

X = iris.iloc[:, 0:4].values
X
```

```python
from sklearn.cluster import KMeans
```

```python
'''
Após isso vamos inicializar o K-means utilizando 3 clusters e utilizar o método de inicialização random que como vimos irá selecionar os centroides de forma aleatória.
random: Se refere ao modo de inicialização de forma aleatória, ou seja, os centróides iniciais serão gerados de forma totalmente aleatória sem um critério para seleção.
'''
kmeans = KMeans(n_clusters = 3, init = 'random')
```

```python
'''
Observe que o que fiz foi declarar o objeto k-means, agora iremos executar o método fit() para executar o algoritmo e agrupar os dados.
O método fit() recebe como parâmetro os dados a serem agrupados, nesse caso será a variável X que declaramos anteriormente.
'''
kmeans.fit(X)
```

```python
'''
Neste momento já temos os dados agrupados e vamos verificar os centroides gerados através do atributo cluster_centers_.
'''
kmeans.cluster_centers_
```

```python
'''
Outra funcionalidade  interessante que vale a pena dar uma olhada é o método fit_transform().
O método fit_transform() executa o K-means para agrupar os dados e retorna uma tabela de distâncias.
A tabela de distâncias é criada de forma que em cada instância contém os valores de distância em relação a cada cluster.
'''
distance = kmeans.fit_transform(X)
distance
```

```python
'''
Verifique que na coluna a direita temos os menores valores…0.14, 0.43, 0.41 e assim por diante, nesse caso significa que essas instâncias tem um menor valor de distância para o respectivo cluster.
Podemos conferir visualizando o atributo labels_ que nos retorna os labels para cada instância, ou seja, o código do cluster que a instância de dados foi atribuído.
'''
labels = kmeans.labels_
labels
```

```python
'''
O método Elbow se trata de uma técnica interessante para encontrar o valor ideal do parâmetro k.
Basicamente o que o método faz é testar a variância dos dados em relação ao número de clusters.
É considerado um valor ideal de k quando o aumento no número de clusters não representa um valor significativo de ganho.
Para ficar mais fácil e claro o entendimento vamos plotar um gráfico e observar o cotovelo (Elbow) se formando.
'''
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
wcss = []
 
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'random')
    kmeans.fit(X)
    print(i, kmeans.inertia_)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('O Metodo Elbow')
plt.xlabel('Numero de Clusters')
plt.ylabel('WSS') #within cluster sum of squares
plt.show()
```

```python
'''
Com o nosso modelo já criado podemos usar o algoritmo K-means para agrupar novos dados baseado nos centroides que já estão definidos.
O processo é extremamente simples..
neste caso, o que será feito é o cálculo de distância entre os novos dados e os centroides de nosso modelo, qual centroide apresentar a menor distância, será o cluster escolhido.
Não precisamos fazer estes cálculos a Scikit-Learn nos fornece essa funcionalidade através do método predict().
'''
data = [
        [ 4.12, 3.4, 1.6, 0.7],
        [ 5.2, 5.8, 5.2, 6.7],
        [ 3.1, 3.5, 3.3, 3.0]
    ]
kmeans.predict(data)
```

```python
'''
Para ficar ainda mais claro é interessante visualizar através de uma representação gráfica os nossos dados e os centroides criados pelo K-means
'''
plt.scatter(X[:, 0], X[:,1], s = 100, c = kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'red',label = 'Centroids')
plt.title('Iris Clusters and Centroids')
plt.xlabel('SepalLength')
plt.ylabel('SepalWidth')
plt.legend()

plt.show()
```
